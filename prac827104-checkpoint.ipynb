# Transient General-Head Boundary Package

Adding in details about the transient general head boundary package. 

<font color='red'> Whats going on here? Comment line by line </font>
    
Relevant documentation https://modflowpy.github.io/flopydoc/mfghb.html

```python
# Make list for stress period 1
stageleft = 10. #the left-most head
stageright = 10. #the right-most head
bound_sp1 = [] #creates an empty vector
for il in range(nlay): #creates a for loop which applies the conductance to the cells for each of the layers
    condleft = hk [il]* (stageleft - zbot/3) * delc #hydraulic conductance from hydraulic conductivity from the left most head to the bottom
    condright = hk [il]* (stageright - zbot/3) * delc #hydraulic conductance from hydraulic conductivity from the right-most head to the bottom
    print (hk)
    for ir in range(nrow): #for loop which applies the conductance to the cells for each row 
        bound_sp1.append([il, ir, 0, stageleft, condleft]) #fills the empty vetor 
        bound_sp1.append([il, ir, ncol - 1, stageright, condright]) #fills the empty vector
print ('Adding ', len(bound_sp1), 'GHBs for stress period 1.')       #displays general head boundaries for stress period 1

# Make list for stress period 2
stageleft = 10. #the left-most head
stageright = 0. #the right-most head
bound_sp2 = []
for il in range(nlay): #creates a for loop which applies the conductance to the cells for each of the layers 
    condleft = hk [il]* (stageleft - zbot) * delc #hydraulic conductance from hydraulic conductivity from the left most head to the bottom
    condright = hk [il]* (stageright - zbot) * delc #hydraulic conductance from hydraulic conductivity from the right-most head to the bottom
    print (hk)
    for ir in range(nrow): #for loop which applies the conductance to the cells for each row 
        bound_sp2.append([il, ir, 0, stageleft, condleft]) #fills the empty vector 
        bound_sp2.append([il, ir, ncol - 1, stageright, condright]) #fills the empty vector 
print ('Adding ', len(bound_sp2), 'GHBs for stress period 2.') #displays general head boundaries for stress period 2

# We do not need to add a dictionary entry for stress period 3.
# Flopy will automatically take the list from stess period 2 and apply it
# to the end of the simulation, if necessary
                         
stress_period_data = {0: bound_sp1, 1: bound_sp2} #creates an array setting the period 1 as the first element, 2 as the second and so on

# Create the flopy ghb object
ghb = flopy.modflow.ModflowGhb(mf, stress_period_data=stress_period_data)
```

```python
#Import the libraries                                
import flopy.utils.binaryfile as bf           #imports flopy as a binary file

# Create the headfile object 
headobj = bf.HeadFile(modelname+'.hds')       #produces 2nd head array with precission 1
times = headobj.get_times()                   #produces a list of unique floating head times

# Setup contour parameters
levels = np.arange(1, 10, 1)  #try changing this             #produces evenly spaced numbers over a 1-10 interval
extent = (delr/2., Lx - delr/2., delc/2., Ly - delc/2.)	     #sets limits of the range of x axis and y axis 
print ('Levels: ', levels)                                   #displays levels values
print ('Extent: ', extent)                                   #displays extent values
# Well point
wpt = ((float(ncol/2)-0.5)*delr, (float(nrow/2-1)+0.5)*delc) #converts the coordinates into floating points
wpt = (450., 550.)                                           #sets the location of well

# Make the plots
mytimes = [1.0, 101.0, 201.0]                                #creates mytimes array with values
for iplot, time in enumerate(mytimes):                       #creates indx list 
    print ('*****Processing time: ', time)                   #displays the processing time
    head = headobj.get_data(totim=time)                      #gets data from the headfile
    #Print statistics
    print ('Head statistics')                                #displays a title for the plot
    print ('  min: ', head.min())                            #displays minimum head value
    print ('  max: ', head.max())                            #displays maximum head value 
    print ('  std: ', head.std())                            #displays standard deviation of values

    #Create the plot
    #plt.subplot(1, len(mytimes), iplot + 1, aspect='equal')
    plt.subplot(1, 1, 1, aspect='equal')                                       #produces a subplot on the figure
    plt.title('stress period ' + str(iplot + 1))                               #displays plot title 
    plt.imshow(head[0, :, :], extent=extent, cmap='BrBG', vmin=0., vmax=10.)   #plots the head data
    plt.colorbar()                                                             #plots the colourbar
    CS = plt.contour(np.flipud(head[0, :, :]), levels=levels, extent=extent,
                     zorder=10)                                                #add contours onto figure
    plt.clabel(CS, inline=1, fontsize=10, fmt='%1.1f', zorder=11)              #add labels onto figure
    mfc = 'None'
    if (iplot+1) == len(mytimes):
        mfc='black'
    plt.plot(wpt[0], wpt[1], lw=0, marker='o', markersize=8,                   #creates plot and selects marker and marker size
             markeredgewidth=0.5,                                              #selects marker edge width
             markeredgecolor='black', markerfacecolor=mfc, zorder=9)           #selects marker edge colour
    plt.text(wpt[0]+25, wpt[1]-25, 'well', size=12, zorder=12)                 #plots well onto well cell
    plt.show()                                                                 #displays plot 

plt.show()  #displays plot

```
